```python
import imageio
import torch
import os

#Load enkelt billed til tensor
#img_arr = imageio.imread('../content/sample_data/bird.jpg')
#img = torch.from_numpy(img_arr)
#out = img.permute(2,0,1)

#Lav tensors
#Parametre er størrelse, channels, højde, bredde og farve-reprenstation (8-bit integer her)
batch_size = 4
batch = torch.zeros(batch_size, 3, 600, 900, dtype=torch.uint8)

#Kig igennem directory og hent alle jpg billede
data_dir = '../content/sample_data/images/'
filenames = [name for name in os.listdir(data_dir)
if os.path.splitext(name)[-1] == '.jpg']
print('file order:')
for i, filename in enumerate(filenames):
  #Konvertere til tensor og smid i batch
  img_arr = imageio.imread(os.path.join(data_dir, filename))
  print(filename)
  img_t = torch.from_numpy(img_arr)
  img_t = img_t.permute(2, 0, 1)
  img_t = img_t[:4]
  batch[i] = img_t

#Lav tensor (RGB) til float, så man kan tage mean
batch = batch.float()
batch /= 255
print('\n')
#gå igennem alle batchs og print mean
print('Tensor channels and brightness:')
for i, c in enumerate(batch):
  print('Tensor ' + str(i) + ':')
  print('Brightness:',torch.mean(c)) #brightness
  #Channel mean, ordered by RGB(I think)
  print('Red:',c[0].mean()*255) #red
  print('Green:',c[1].mean()*255) #blue
  print('Blue:',c[2].mean()*255) #green
  print('\n')

```
