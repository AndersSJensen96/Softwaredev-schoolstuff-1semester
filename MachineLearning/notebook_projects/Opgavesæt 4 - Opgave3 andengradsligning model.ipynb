```python
import torch


x_values = torch.tensor([-2, -1.5, -1, -0.5, 0, 0.5, 1, 1.5, 2, 2.5, 3, 3.5])

y_values = torch.tensor([-17, -9.75, -4, 0.25, 3, 4.25, 4, 2.25, -1, -5.75, -12, -19.75])

#change the number after minus to get different loss
learning_rate = 1e-5

# weights = torch.ones(())
# biases = torch.zeros(())
# c_values = torch.zeros(())

#Model as in the function, returns predictions
def model(x, a, b, c) :
  return a*x**2+b*x+c                     

#Loss function, used to calculate the loss rate
def loss_func(prediction, actual_value) :
  squared_diff = (prediction - actual_value)**2
  return squared_diff.mean()

#derivative of lossfunction x^2 (there exists more loss functions to get the difference)
def derivative_Loss_func(prediction, actual_value) :
  # don't quite understand the division, dividere m√•ske pga det er et gennemsnit man f√•r med lossfunc'(x)
  return 2*(prediction-actual_value) / prediction.size(0)

def derivative_a(x_value):
  derived_a = x_value**2
  return derived_a

#These seem pointless to make, but whatever
def derivative_b(x_value):
  return x_value

def derivative_c():
  return 1

def gradient_function(x_value, prediction) :
  #rule of chain
  dloss_prediction = derivative_Loss_func(prediction, x_values)
  dloss_a = derivative_a(x_value) * dloss_prediction
  dloss_b = derivative_b(x_value) * dloss_prediction
  dloss_c = derivative_c() * dloss_prediction

  return torch.stack([dloss_a.sum(), dloss_b.sum(), dloss_c.sum()])

#Epoch = amount of loops
def training_loop(n_epochs, learning_rate, params, x_value, y_value):
  for epoch in range(1, n_epochs+1) :

    #Set autograd to zero if it's none
    if params.grad is not None:
      params.grad.zero_()

    a,b,c = params
    prediction = model(x_value, *params) #forward, #*params was a,b,c before autograd
    loss = loss_func(prediction, y_value)

    loss.backward() #autograd true
    #gradient = gradient_function(x_value, prediction) #backward, manual gradient



    #just√©r params (a,b,c)
    #autograd true
    with torch.no_grad():
      params -= learning_rate * params.grad
    #manual grad
    #params = params - learning_rate*gradient

    print('Epoch: %d, Loss: %f' % (epoch, float(loss)))
    # print('predictions:')
    # print(prediction)

  return params

#params er tilf√¶ldige, √¶ndre epoch for at f√• mere pr√¶cis v√¶rdier, Allan k√∏rte vidst 600k
training_loop(10000, learning_rate, torch.tensor([1.0, 0.0, 1.0], requires_grad=True), x_values, y_values)

```
